```python
import pandas as pd

#환율 2021.07.15일자
fx_url = 'https://www.ibk.co.kr/fxtr/excRateList.ibk' + '?indate=' + '2021.07.15'
fx_df = pd.read_html(fx_url)[0]
```

```python
#현찰살때,현찰팔때 숫자에 ',' 제거
fx_df['현찰살때(스프레드)'] = fx_df['현찰살때(스프레드)'].str.replace(',','')

fx_df['현찰팔때(스프레드)'] = fx_df['현찰팔때(스프레드)'].str.replace(',','')
```

```python
#regular expression == regex(정규표현식) -> 따로 공부해야함
# [\.\d]+ - > 000.000(~%)에서 000.000 데이터만 추출하고자 함, +를 붙여야 전체가 나옴
# expand = True -> 데이터 프레임
# expand = False -> 시리즈 형태

fx_df['현찰살때(스프레드)'].str.extract(r'([\.\d]+)', expand = False)
```

```python
fx_df['현찰살때'] = fx_df['현찰살때(스프레드)'].str.extract(r'([\.\d]+)', expand = False)
fx_df['현찰팔때'] = fx_df['현찰팔때(스프레드)'].str.extract(r'([\.\d]+)', expand = False)

fx_df
```

```python
#'((.*?))%\)' -> 스프레드 값에서 %제거해 추출
fx_df['현찰살때(스프레드)'].str.extract('\((.*?)%\)', expand = False)
```

```python
fx_df['현찰살때_스프레드'] = fx_df['현찰살때(스프레드)'].str.extract('\((.*?)%\)', expand = False)
fx_df['현찰팔때_스프레드'] = fx_df['현찰팔때(스프레드)'].str.extract('\((.*?)%\)', expand = False)

fx_df
```

```python
fx_df = fx_df.drop(['현찰살때(스프레드)', '현찰팔때(스프레드)'], axis=1)
```
